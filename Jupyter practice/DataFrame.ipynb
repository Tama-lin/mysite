# 組裝自己的DataFrame

```python
import numpy as np
import pandas as pd
```

```python
mydata = np.random.randn(4,3)
df1 = pd.DataFrame(mydata, columns = list("ABC")) # creat a dataframe by using mydata and adding columns of A,B,C
df1
```

```python
df2 = pd.DataFrame(np.random.randn(3,3), columns = list("ABC"))
# creat a dataframe of 3*3 array and adding columns of A,B,C

df2
```

# Concatenate dataframes

```python
df3 = pd.concat([df1,df2], axis = 0) # put df1 and df2 together as a connected series(Merge up and down)
df3
```

```python
df3.index = range(1,8) # change df3 index to range 1 to 7
df3
```

```python
df4 = pd.concat([df1,df2], axis = 1) # put df1 and df2 together as a connected series(Merge left and right)
df4
```

# Change dataframe columns

```python
df4.columns = list('ABCDEF')
df4
```

# Change dataframe index

```python
df4.index = range(1,5)
df4
```

```python
df4.index = list('甲乙丙丁')
df4
```

# Pandas 的 loc 定位法
找到你要的資料，更進一步更改資料

```python
df5= pd.DataFrame(np.random.randn(5,3), index = list(range(1,6)), columns = list('ABC'))
df5
```

```python
df5.loc[2:3, "B":"C"]
```

```python
df5.loc[5, "B"]
```

```python
df5.loc[5, "B"] = 10 # Change data at row 5 coloum B to value 10
df5
```

```python
df5.loc[df5.B > 0, "C"] = 0 # Change data at coloum C which row B >0   to value 0
df5
```

```python
df5.loc[ 3:5, "B":"C"] = 0 
df5
```
