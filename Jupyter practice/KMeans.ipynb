# K-Mean
### Unsupervised learning 自動分類學習
不知道數據如何分類時，比如收集大量的客戶資料時以K-Mean自動分類後，在考察每一類有什麼特性。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

# Prepocessing

```python
X = np.random.rand(100, 2)
X[0]
```

```python
plt.scatter(X[:,0], X[:,1], s = 50)
```

# Machine Learning

```python
clf = KMeans(n_clusters = 3)
# 開Ｋ-Mean機器的方法：機器變數名＝KMeans(n_clusters = 分幾類) 
```

```python
clf.fit(X) 
```

```python
clf.labels_ # Check labels
```

```python
plt.scatter(X[:,0], X[:,1], c = clf.labels_)
```
